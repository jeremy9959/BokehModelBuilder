```python
def prop_show(obj):
    for x in obj.properties_with_refs():
        print(x)
        try:
            print(x,getattr(obj,x).references())
        except AttributeError:
            continue
```

```python
def recur_show(obj):
    for x in obj.properties_with_refs():
        if not isinstance(getattr(obj,x),list):
            try:
                print(x,getattr(obj,x).references())
            except AttributeError:
                continue
        else:
            if len(getattr(obj,x))>0:
                print(x,end="\t")
            for z in getattr(obj,x):
                print(z,end="\t")
                recur_show(z)
            
```

```python
import numpy as np

from bokeh.io import curdoc, show, output_notebook
from bokeh.models import ColumnDataSource, Grid, LinearAxis, Plot, Square
output_notebook()
N = 9
x = np.linspace(-2, 2, N)
y = x**2
sizes = np.linspace(10, 20, N)

source = ColumnDataSource(dict(x=x, y=y, sizes=sizes))

plot = Plot(
    title=None, plot_width=300, plot_height=300,
    min_border=0, toolbar_location=None)

glyph = Square(x="x", y="y", size="sizes", fill_color="#74add1")
plot.add_glyph(source, glyph)

xaxis = LinearAxis()
plot.add_layout(xaxis, 'below')

yaxis = LinearAxis()
plot.add_layout(yaxis, 'left')

plot.add_layout(Grid(dimension=0, ticker=xaxis.ticker))
plot.add_layout(Grid(dimension=1, ticker=yaxis.ticker))

#curdoc().add_root(plot)

show(plot)
```
