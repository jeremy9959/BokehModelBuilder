```python
from bokeh.plotting import figure, Figure
from bokeh.io import output_notebook, show
import numpy as np
output_notebook()
```

```python
F = figure(title='Sin and Cos Curve')
```

```python
F.line(x=np.arange(-10,10,.01),y=np.sin(np.arange(-10,10,.01)),line_width=4,legend_label='sin')
F.line(x=np.arange(-10,10,.01),y=np.cos(np.arange(-10,10,.01)),line_width=4,color='orange',legend_label='cos')
```

```python
show(F)
```

```python
from bokeh.models import LinearAxis
import pandas as pd
```

```python
def obj_props_to_df2(obj):
    tp = type(obj)
    obj_dict = obj.properties_with_values()
    types = [obj.lookup(x) for x in obj_dict.keys()]
    docs = [getattr(type(obj),x).__doc__ for x in obj_dict.keys()]
    return pd.DataFrame.from_dict({'props':list(obj_dict.keys()), 'values':list(obj_dict.values()),'types':types,'doc':docs})
```

```python
obj_props_to_df2(LinearAxis())
```
