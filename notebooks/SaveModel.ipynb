```python
import bokeh
import numpy as np
from bokeh.document import Document
from bokeh.plotting import figure
from bokeh.io import output_notebook, show, curdoc, push_notebook
from bokeh.layouts import row
import json
output_notebook()

F = figure()
F.line(list(range(-10,10)),list(range(-10,10)))
F.scatter(np.random.normal(0,5,size=100),np.random.normal(0,5,size=100))
show(F)

d=curdoc().to_json_string()
with open('document.json','w') as f:
    f.write(d)
```
