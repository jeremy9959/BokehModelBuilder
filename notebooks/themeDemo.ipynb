```python
from bokeh.plotting import figure, output_notebook, show
from bokeh.models import BoxSelectTool, Toolbar
from bokeh.themes import Theme
from bokeh.io import curdoc
```

```python
output_notebook()
```

```python
x = [1, 2, 3, 4, 5]
y = [6, 7, 6, 4, 5]

theme = Theme('valentines.yaml')
curdoc().theme = theme
p = figure(title='Valentines Day', plot_width=400, plot_height=400)
p.line(x, y,color='yellow',line_width=5)
show(p)
```
